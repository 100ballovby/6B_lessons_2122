## Алгоритм линейного поиска элемента в массиве

<p><b>Линейный поиск</b> - это базовый алгоритм поиска значения в коллекции, который заключается в последоовательном переборе каждого элемента коллекции и сравнения его с искомым значением. Если значение найдено, поиск должен быть остановлен.</p>

```python
lst = [5, 4, 13, 8, 0, 2, 98, 57, 13, 21, 6]
```

<p>Предположим, что я хочу найти какое-то число в этом списке. Для этого я создам переменную, в которую пользователь будет вводить искомое значение - <code>key</code>.</p>

```python
key = int(input('Введите число: '))
```

<p>Теперь моя задача просмотреть каждый элемент списка (цикл <code>for</code>, количество повтоорений которого равно <code>len(lst)</code>) и сравнить каждый элемент с переменной <code>key</code>. Если элемент и переменная равны (<code>==</code>), то поиск останавливается конструкцией <code>break</code> и выводится индекс найденного элемента. Если элемент не был найден, нужно вывести сообщение <i><b>"Такого элемента нет"</b></i></p>

```python
for index in range(len(lst)):
    if lst[index] == key:
        break
        
if lst[index] == key:
    print(f'id = {index}')
else:
    print('Элемент не найден!')
```

<p>АЛП является самым медленным алгоритмом поиска. Время исполнения такого поиска <b>линейно</b>, то есть количество операций зависит от количества элементов последовательности.</p>

<p>Мы рассматриваем всегда самый худший случай, когда искомое число находится в самом конце списка. Соответственно, не сложно посчитать, что для массива длиной 500 элементов, нужно совершить 499 попыток.</p>

### Другие применения

<p>Использовать модифицированный алгоритм линейного поиска можно в обычных задачах на списки.</p>

<p>Предположим, что я хочу заменить все нулевые элементы списка значением <code>None</code>. Для этого я должен просмотреть каждый элемент и сравнить его с нулем. То есть по факту, это тот же самый АЛП, но он не возвращает индекс найденного элемента, а просто помогает его обнаружить.</p>

```python
lst2 = [5, 0, 12, 0, 7, 8, 9, 0, 3, 5, 0, 1, -8]
# нужно заменить все нули значением None 

for index in range(len(lst2)):
    if lst2[index] == 0:
        print(f'Заменяю число с индексом {index}')
        lst2[index] = None  # переназначаю элемент
        
print(f'Преобразованный: {lst2}')
```
